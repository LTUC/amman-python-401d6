```
## Count the number of False to True transitions
```

```
import numpy as np
```

```
bools = np.random.choice([True, False], size=10000)
```

```
bools[:20]
```

```
np.bincount(bools[:20])
```

```
np.count_nonzero(bools[:20])
```

```
def count_transitions(bools):
  count = 0
  for a,b in zip(bools[:-1],bools[1:]):
    if b and not a:
    # if b == True and a == False: # NOTE: compare approaches when you get to timing step
      count += 1
  return count
```

```
count_transitions(bools)
```

```
np.count_nonzero(bools[:-1] < bools[1:])
```

```
# profile it
from timeit import timeit
setup = 'from __main__ import count_transitions, bools; import numpy as np'
num = 1000
t1 = timeit('count_transitions(bools)', setup=setup, number=num)
t2 = timeit('np.count_nonzero(bools[:-1] < bools[1:])', setup=setup, number=num)
print('Speed difference: {:0.1f}x'.format(t1 / t2))
```

## Buy Low Sell High

```
def profit(prices):
  max_px = 0
  min_px = prices[0]
  for px in prices[1:]:
    min_px = min(px, min_px)
    max_px = max(px - min_px, max_px)
  return max_px
```

```
prices = (20, 18, 14, 17, 20, 21, 15)
profit(prices)
```

```
cummin = np.minimum.accumulate

def profit_with_numpy(prices):
  prices = np.asarray(prices)
  return np.max(prices - cummin(prices))
```

```
prices = (10, 15, 20, 18, 14, 17, 20, 21, 6, 15)
profit_with_numpy(prices)
```
