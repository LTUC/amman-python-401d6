```
def is_diagonal(ax,ay,bx,by):
  """
  given coordinates for 2 points a and b
  determine if they are diagonals
  """
  dx = abs(ax - bx)
  dy = abs(ay - by)
  return dx == dy
```

```
# tests
assert is_diagonal(0,0,1,1) == True
assert is_diagonal(0,0,1,2) == False
assert is_diagonal(3,0,2,1) == True
assert is_diagonal(1,1,0,0) == True
assert is_diagonal(1,2,0,0) == False
assert is_diagonal(2,1,3,0) == True
assert is_diagonal(5,4, 99,98)

print('........ CONGRATULATIONS : ALL TESTS PASS ........')
```
