## Make an 3 by 3 array filled with 3s in 3 different ways

```
import numpy as np
```

```
np.full((3,3),3)
```

```
np.ones((3,3),dtype=np.int) * 3
```

```
np.zeros((3,3), dtype=np.int) + 3
```

```
arr = np.array([[1,2,3],[4,5,6]])
np.isin(7, arr)
```

```
def is_needle_in_haystack(needle, haystack):
  return np.isin(needle, haystack)
```

```
needle = 6
haystack = np.zeros((3,3,3))
haystack[1][1][1] = needle

found_it = is_needle_in_haystack(needle, haystack)

print(found_it)
```

```
nums = np.random.randint(100,size=100)
```

```
nums.cumsum()
```

```
# from iter
singles = range(10)
siter = iter(singles)
# next(siter) # checkout out how nums changes when advancing iterator
nums = np.fromiter(siter,dtype=np.int)
nums
```

```
# arange
np.arange(10,20,2)
```

```
# linspace
np.linspace(10,20,5, retstep=True)
```

```
a = np.arange(10) 
print(a)
s = slice(2,7,2) 
print(a[s])
print(a[2:7:2])
```

```
# indexing with ...
a = np.array([[1,2,3],[3,4,5],[4,5,6]]) 
a[...,2]
```
