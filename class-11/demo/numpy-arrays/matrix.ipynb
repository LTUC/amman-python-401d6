```
import numpy as np
```

```
mat = np.zeros((8,8), dtype="int")
print(mat)
```

```
mat[3][6] = 1
mat
```

```
# 1 in mat[6]
# 1 in mat.transpose()[6]
```

```
def in_column(grid, col, value):
  return value in grid.transpose()[col]
```

```
# tests
success = '........ ALL TESTS PASSED ........'
assert in_column(mat, 6, 1) == True
print(success)
```
