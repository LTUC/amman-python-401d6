```python
def manufacturer_name(mfr):
  map = {
    "A" : "American Home Food Products;",
    "G" : "General Mills",
    "K" : "Kelloggs",
    "N" : "Nabisco",
    "P" : "Post",
    "Q" : "Quaker Oats",
    "R" : "Ralston Purina",
  }
  return map.get(mfr)

df['Manufacturer'] = df['mfr'].apply(manufacturer_name)
df[['mfr','Manufacturer']]
```
